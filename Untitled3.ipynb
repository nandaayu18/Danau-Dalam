```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as scp
```

```python
lakes = pd.read_csv('List of lakes.csv')
```

```python
lakes.dtypes
```

```python
#mencari data yang mising value
lakes.isna()
```

```python
#membersihkan atau clean data yang mengalami mising value
lakes = lakes.dropna()
lakes.head()
```

```python
lakes
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = (lakes.Name) 
y = (lakes.Country)
ax.bar(x,y)
plt.xticks (rotation = 'vertical')
plt.xlabel("Nama Danau")
plt.ylabel("Negara")
plt.show()
```

```python
def countplot(column):
    return sns.countplot(x= column, data= lakes)
```

```python
countplot('Country')
plt.xticks (rotation = 'vertical')
```
